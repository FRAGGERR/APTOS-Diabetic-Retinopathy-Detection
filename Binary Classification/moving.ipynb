```python
import os
import shutil

# Define the source and destination directories
src_dir = '/Users/hardikchhipa/Desktop/DR-Arvind/Binary Classification on Retina/dataset/ImageNet-Mini/images'
dst_dir = '/Users/hardikchhipa/Desktop/DR-Arvind/Binary Classification on Retina/dataset/non-dr'

# Create the destination folder if it doesn't exist
os.makedirs(dst_dir, exist_ok=True)

# Log the process
moved_files_count = 0

# Loop over all subdirectories and move the images
for subdir, _, files in os.walk(src_dir):
    for file in files:
        if file.lower().endswith(('.png', '.jpg', '.jpeg', '.bmp', '.tiff')):
            src_file = os.path.join(subdir, file)
            dst_file = os.path.join(dst_dir, file)

            # Handle filename conflict by adding a suffix if the file already exists
            if os.path.exists(dst_file):
                base, ext = os.path.splitext(file)
                i = 1
                while os.path.exists(dst_file):
                    dst_file = os.path.join(dst_dir, f"{base}_{i}{ext}")
                    i += 1

            # Move the file
            shutil.move(src_file, dst_file)
            moved_files_count += 1
            print(f"Moved: {src_file} to {dst_file}")

if moved_files_count > 0:
    print(f"All {moved_files_count} images have been moved successfully!")
else:
    print("No images were found or moved.")
```
