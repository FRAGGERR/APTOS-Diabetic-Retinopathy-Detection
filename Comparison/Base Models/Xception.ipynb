```python
import numpy as np
import matplotlib.pyplot as plt

# Example data
metrics = ['Accuracy', 'Precision', 'Recall', 'F1 Score']

# Data for the previous dataset
previous_train = [89.35, 91.58, 87.01, 89.18] 
previous_val = [73.10, 78.37, 71.46, 72.89]

# Data for the APTOS dataset
aptos_train = [91.23, 93.20, 89.33, 90.29]
aptos_val = [78.57, 80.75, 77.34, 78.85] 

# Bar width and positions
bar_width = 0.23
index = np.arange(len(metrics))

# Set the same color scheme for both datasets
train_color = '#0000FF'  # Blue
val_color = '#FFA500'    # Orange

# Plot bars
plt.bar(index, previous_train, bar_width, label='Previous - Training', color=train_color)
plt.bar(index + bar_width, previous_val, bar_width, label='Previous - Validation', color=val_color)

# Add APTOS dataset bars using the same color scheme
plt.bar(index, aptos_train, bar_width, alpha=0.5, color=train_color, label='APTOS - Training')
plt.bar(index + bar_width, aptos_val, bar_width, alpha=0.5, color=val_color, label='APTOS - Validation')

# Labels and title
plt.xlabel('Metrics')
plt.ylabel('Percentage')
plt.title('Model Performance Metrics - Xception')
plt.xticks(index + bar_width / 2, metrics)
plt.legend()
plt.ylim([60, 100])

# Show plot
plt.tight_layout()
plt.show()
```
